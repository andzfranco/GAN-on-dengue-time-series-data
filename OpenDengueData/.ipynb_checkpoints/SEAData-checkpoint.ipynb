```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = '/Users/andrearosefranco/VIP/AY2024-2025/OpenDengueData/Philippines.csv'
data = pd.read_csv(file, header=0)

data = data[data['T_res'] == 'Month']
data_cases = data[['calendar_end_date', 'dengue_total']]
data_cases['calendar_end_date'] = pd.to_datetime(data_cases['calendar_end_date'])
data_cases.set_index('calendar_end_date', inplace=True)

date_range = pd.date_range(start=data_cases.index.min(), end=data_cases.index.max(), freq='M')
data_cases = data_cases.reindex(date_range)
#data_cases.fillna(value=np.nan, inplace=True)

display(data_cases)
plt.plot(data_cases['dengue_total'])
plt.title('Monthly Dengue Cases in Philippines')
plt.ylabel('Dengue Cases')
plt.xlabel('Year')
plt.show()
```

# Get 2D histogram

```python
def get_sparsity(series):
    total_data = len(series)
    missing_data = series.isnull().sum()
    percent_sparse = (missing_data / total_data) * 100
    print(percent_sparse)
    return percent_sparse

def get_gaplengths(series):
    gaplengths = (series.isnull().astype(int)
                   .groupby(series.notnull().cumsum()).sum().values)
    return gaplengths[gaplengths > 0] 

def plot_sparsity_vs_gap_lengths(df, columns):
    sparsity_ratios = []

    for col in columns:
        series = df[col]
        sparsity = get_sparsity(series)
        sparsity_ratios.append(sparsity)

        gaplengths = (get_gaplengths(series))
        print(type(gaplengths))
        sparsity_ratios.extend([sparsity] * (len(gaplengths)-1))
        print(type(sparsity_ratios))

    plt.figure(figsize=(8, 6))
    plt.hist2d(sparsity_ratios, gaplengths, bins=20, cmap='Blues')
    plt.colorbar(label='Frequency')
    plt.xlabel('Sparsity (%)')
    plt.ylabel('Gap Length')
    plt.title('2D Histogram of Sparsity vs Gap Lengths')
    plt.show()
```

```python
plot_sparsity_vs_gap_lengths(data_cases, ['dengue_total'])
```
