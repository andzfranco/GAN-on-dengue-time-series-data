```python
def generate_sparse_data(time_steps, sparsity_ratio):
    clean_data = np.sin(np.linspace(0, 100, time_steps)) #create sine wave
    noise = np.random.normal(0, 0.1, clean_data.shape)  #mean 0, standard deviation 0.1
    ground_truth_data = clean_data + noise
    num_sparse_data = int(time_steps * sparsity_ratio)
    print(num_sparse_data)
    missing_indices = np.random.choice(range(1, time_steps), num_sparse_data, replace=False) #randomly get indices from data
    sparse_data = ground_truth_data.copy()
    sparse_data[missing_indices] = np.nan #replace these data with nan
    return ground_truth_data, sparse_data

def build_rnn_impute():
    model = tf.keras.Sequential([
        layers.Input(shape=(None, 1)),
        layers.SimpleRNN(50, activation='relu', return_sequences=False),
        layers.Dense(1)])
    return model

def build_rnn():
    model = tf.keras.Sequential([
        layers.Input(shape=(None, 1)),
        layers.SimpleRNN(50, activation='relu', return_sequences=False),
        layers.Dense(latent_dim)])
    return model

def build_generator():
    noise_input = layers.Input(shape=(10,))
    condition_input = layers.Input(shape=(latent_dim,))
    merged_input = layers.concatenate([noise_input, condition_input])
    x = layers.Dense(128, activation='relu')(merged_input)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dense(time_steps, activation='tanh')(x)
    model = tf.keras.Model([noise_input, condition_input], x)
    return model

def build_discriminator():
    data_input = layers.Input(shape=(time_steps,))
    condition_input = layers.Input(shape=(latent_dim,))
    merged_input = layers.concatenate([data_input, condition_input])
    x = layers.Dense(256, activation='relu')(merged_input)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dense(1, activation='sigmoid')(x)
    model = tf.keras.Model([data_input, condition_input], x)
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model

def build_gan(generator, discriminator):
    discriminator.trainable = False
    noise_input = layers.Input(shape=(10,))
    condition_input = layers.Input(shape=(latent_dim,))
    generated_data = generator([noise_input, condition_input])
    gan_output = discriminator([generated_data, condition_input])
    gan = tf.keras.Model([noise_input, condition_input], gan_output)
    gan.compile(optimizer='adam', loss='binary_crossentropy')
    return gan

def plot_real_and_generated(real_data, generated_data, sparse_data, epoch, sparsity_ratio):
    plt.figure(figsize=(8, 4))
    plt.plot(real_data, label="Real Data", alpha=0.6)
    plt.plot(generated_data, label="Generated Data", alpha=0.6)
    plt.plot(sparse_data, label="Sparse Data", linestyle='--', color='r', alpha=0.6)
    plt.title(f'Sparse ({sparsity_ratio}%){i} vs Generated Data at Epoch {epoch}')
    plt.legend()
    plt.show()

def train_cgan(gan, generator, discriminator, rnn, data, epochs):
    sparse_data = data.copy()
    plt.figure(figsize=(8,4))
    plt.plot(sparse_data)
    plt.title('Sparse data')
    
    #Impute missing values using the RNN 
    rnn_impute = build_rnn_impute() 
    for i in range(time_steps):
        if np.isnan(data[i]):
            rnn_input = np.expand_dims(data[:i], axis=0)
            rnn_output = rnn_impute.predict(rnn_input, verbose=0)
            data[i] = rnn_output
    
    #For plotting of loss
    d_losses = []
    g_losses = []
    
    #Train the GAN on the fully imputed data
    for epoch in range(epochs):
        noise = np.random.normal(0, 1, (1, 10))  #Generate noise

        real_labels = np.ones((1, 1)) 
        fake_labels = np.zeros((1, 1))  
        condition = rnn.predict(np.expand_dims(data, axis=0), verbose=0)[0]   

        #Train discriminator to idenfity real data, real label
        d_loss_real = discriminator.train_on_batch([np.expand_dims(data, axis=0), np.expand_dims(condition, axis=0)], real_labels)
        
        #Generate fake data with the generator
        generated_data = generator.predict([noise, np.expand_dims(condition, axis=0)], verbose=0)
        
        #Train the discriminator on the generated fake data, fake label
        d_loss_fake = discriminator.train_on_batch([generated_data, np.expand_dims(condition, axis=0)], fake_labels)
        
        d_losses.append(d_loss_real + d_loss_fake) #Total discrimniator loss, to ensure can identify both fake and real 
        
        #Train the GAN model
        valid_y = np.ones((1, 1)) 
        g_loss = gan.train_on_batch([noise, np.expand_dims(condition, axis=0)], valid_y)
        g_losses.append(g_loss)
        
        if epoch % 200 == 0:
            print(f"EPOCH: {epoch} [D loss: {d_loss_real + d_loss_fake}] [G loss: {g_loss}]")
            plot_real_and_generated(ground_truth_data, generated_data[0], sparse_data, epoch, sparsity_ratio)
            
    return d_losses, g_losses, generated_data[0]
```
