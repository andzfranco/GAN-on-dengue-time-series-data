# Get 2D histogram

```python
def get_sparsity(series):
    total_data = len(series)
    missing_data = series.isnull().sum()
    percent_sparse = (missing_data / total_data) * 100
    return percent_sparse

def get_gaplengths(series):
    gaplengths = (series.isnull().astype(int)
                   .groupby(series.notnull().cumsum()).sum().values)
    return gaplengths[gaplengths > 0] 
```

```python
file = '/Users/andrearosefranco/VIP/AY2024-2025/OpenDengueData/AllData.csv'
data = pd.read_csv(file, header=0)
data = data[data['T_res'] == 'Month']

all_sparsity_ratios = []
all_gap_lengths = []
all_data_lengths = []

for country, group in data.groupby('adm_0_name'):
    #Get date and dengue case columns
    data_cases = group[['calendar_end_date', 'dengue_total']]
    
    #Convert the date column to datetime and set it as the index
    data_cases['calendar_end_date'] = pd.to_datetime(data_cases['calendar_end_date'])
    data_cases.set_index('calendar_end_date', inplace=True)
    date_range = pd.date_range(start=data_cases.index.min(), end=data_cases.index.max(), freq='M')
    data_cases = data_cases.reindex(date_range)
    data_length = len(data_cases) 
    
    #Plotting
    plt.plot(data_cases['dengue_total'])
    plt.title(f'Monthly Dengue Cases in {country}')
    plt.ylabel('Dengue Cases')
    plt.xlabel('Year')
    plt.show()
    
    sparsity = get_sparsity(data_cases['dengue_total'])
    gap_lengths = get_gaplengths(data_cases['dengue_total'])
    print(f'{country}: {sparsity}, {gap_lengths}')
    all_sparsity_ratios.extend([sparsity] * len(gap_lengths)) 
    all_gap_lengths.extend(gap_lengths) 
    all_data_lengths.extend([data_length] * len(gap_lengths))
```

# Fitting

```python
from scipy.optimize import curve_fit
```

```python
start = 0
end = 45
spacing = 5
```

### **Quadratic Regression**

Parabolic best-fit curve

where $R^2 = 1 - \frac{SSR}{SST}$

The sum squared regression (SSR) is the sum of the residuals squared, and the total sum of squares (SST) is the sum of the distance the data is away from the mean all squared.

```python
degree = 2
model = np.poly1d(np.polyfit(all_sparsity_ratios, all_gap_lengths, degree))
polyline = np.linspace(min(all_sparsity_ratios), max(all_sparsity_ratios), 100) 

#Get R^2
predicted_values = model(all_sparsity_ratios)
ss_residual = np.sum((all_gap_lengths - predicted_values) ** 2)
ss_total = np.sum((all_gap_lengths - np.mean(all_gap_lengths)) ** 2)
r_squared = 1 - (ss_residual / ss_total)

#Equation
equation = f"y = {model[0]:.2f}x + {model[1]:.2f}x + {model[2]:.2f}\n$R^2$ = {r_squared:.2f}"

#Plotting
plt.scatter(all_sparsity_ratios, all_gap_lengths, color='red', label='Data points')
plt.plot(polyline, model(polyline), linestyle='--', label='Quadratic Fit')

plt.text(0.05, 0.75, equation, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.xlabel('Sparsity (%)')
plt.ylabel('Gap Length')
plt.legend()
plt.show() 
```

```python
rmse = np.sqrt(np.mean((all_gap_lengths - predicted_values) ** 2))
print(f'RMSE: {rmse}')

#Values
x_values = np.arange(start, end, spacing)
y_values = model(x_values).astype(int)

df = pd.DataFrame({'Sparsity': x_values, 'Gaplength': y_values})
df.to_csv(f'sparsegaplist_fitted_{degree}.csv', index=False)
df
```

## Lorentzian fitting

```python
def lorentzian(x, amp, x0, gamma):
    return amp / (np.pi * gamma * (1 + ((x - x0) / gamma)**2))
```

```python
plt.hist(all_sparsity_ratios, bins=bins, density=True, color='lightseagreen')

#lorentzian
hist, bin_edges = np.histogram(all_sparsity_ratios, bins=bins, density=True)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
initial_guesses = [np.max(hist), np.median(all_sparsity_ratios), (bin_edges[-1] - bin_edges[0]) / 10]
hist, bin_edges = np.histogram(all_sparsity_ratios, bins=bins, density=True)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
popt, pcov = curve_fit(lorentzian, bin_centers, hist, p0 = initial_guesses)
x_fit = np.linspace(bin_edges[0], bin_edges[-1], 1000)
y_fit = lorentzian(x_fit, *popt)
plt.plot(x_fit, y_fit, color="red", label="Lorentzian", linewidth=2)
#source: https://www.statology.org/how-to-use-the-cauchy-distribution-in-python/

plt.xlabel('Missing data (%)')
plt.ylabel('Density')
#plt.title('Sparsity Histogram')
plt.legend(loc='upper left')
plt.tight_layout()
plt.savefig('Lorentzian.png')
plt.show()
```

```python
predicted_values = lorentzian(bin_centers, *popt)
residuals = hist - predicted_values
rmse = np.sqrt(np.mean(residuals**2))
print(rmse)
plt.bar(bin_centers, residuals, width=(bin_edges[1] - bin_edges[0]), color='orange', edgecolor='black')
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.xlabel('Sparsity (%)')
plt.ylabel('Residuals')
plt.title('Residuals of Lorentzian Fit')
plt.tight_layout()
plt.savefig('Residuals_Lorentzian.png')
plt.show()
```
